```python
from cassandra.cluster import Cluster
import redis
```

### CONECTAR A CASSANDRA

```python
cluster = Cluster(['127.0.0.1'], port=9042)   # Cambia por tu IP
session = cluster.connect('demo')  # Tu keyspace
```

### CONECTAR A REDIS

```python
query = "SELECT id, nombre FROM usuarios;"   # Cambia por tus columnas
rows = session.execute(query)
```

```python
r = redis.Redis(host='127.0.0.1', port=27017, db=0)
```

### GUARDAR CADA PAR CHAVE-VALOR EN REDIS

```python
for row in rows:
    key = str(row.id)         # clave
    value = row.nombre        # valor
    r.set(key, value)
    print(f"Guardado en Redis → {key}: {value}")
```

### VERIFICAR DATOS EN REDIS

```python
print("\nVerificación en Redis:")
for test_key in r.keys():
    print(test_key.decode(), "→", r.get(test_key).decode())
```
